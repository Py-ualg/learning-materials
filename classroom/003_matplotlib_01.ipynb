```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
```

### First we create random data

```python
# numerical data
x = np.linspace(0, 2*np.pi, 101)
y = np.sin(x)

# categorical
fruits = ['orange']*2 + ['apple']*4 + ['banana']
```

## Many plots to get you started with

```python
# line plot
plt.plot(y, x)
plt.show()
```

![image](figs/plt_anatomy.png)

```python
# scatter plot
plt.scatter(x, y, marker='s', color='g')
plt.plot(x, y+0.2, marker='s', color='b')
plt.plot(x, y+0.4, 'sr')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
plt.bar(x, y)   # try barh, width, bottom, align
```

```python
delta = 0.025
x1 = y1 = np.arange(-3.0, 3.0, delta)
X, Y = np.meshgrid(x1, y1)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2

plt.imshow(Z,
           # origin='lower', extent=[-3, 3, -3, 3],
           # cmap = cm.Reds_r,
          )
plt.colorbar()
```

```python
## try contourf method too
plt.contour(Z,
            # levels=21,
           )
plt.show()
```

```python
plt.boxplot([x, y])
# plt.boxplot(y)
plt.show()
```

```python
plt.violinplot([x, y],
               showmeans=False,
               showmedians=True,
              )
# plt.boxplot(y)
plt.show()
```

```python
# pie
plt.pie([20, 30, 40, 10])
plt.show()
```

```python
plt.hist(fruits)
plt.show()
```

## Here is the object oriented approach (CONFUSING)
* ax is a plot axis, python object
* allows more customization
* same methods have similar but different names 

```python
# Creating a figure and axes
fig, ax = plt.subplots()

# Plotting data on the axes
ax.plot(x, y, 'r--')  # Red dashed line
ax.scatter(x[::10], y[::10], color='b')  # Blue scatter points

# Customizing the axes labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Figure with Axes')

# Displaying the plot
plt.show()
```

## Multiple plots

```python
fig, ax = plt.subplots(3, 2, figsize=(8, 5))  # sharey, sharex
ax[0, 0].plot(x, np.sin(x))
plt.show()
```
